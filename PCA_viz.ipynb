```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import plotly.express as px
```

```python
features = pd.read_csv("res/features.csv")
features.set_index("MAL_ID", inplace=True)
features.head()
```

```python
target = pd.read_csv("res/popularity.csv")
target.set_index("MAL_ID", inplace=True)
target
```

```python
#PCA Visualization
pca = PCA(n_components= 2, random_state=3000)
features_standardized = StandardScaler().fit_transform(features)
reduced_data = pca.fit_transform(features_standardized)
reduced_df = pd.DataFrame(reduced_data, columns = ["Component1", "Component2"])
reduced_df["target"] = target["Popularity"]

reduced_df
```

```python
graph = px.scatter(reduced_df, x="Component1", y="Component2", color = "target")
graph.show()
```
